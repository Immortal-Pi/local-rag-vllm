```python
from typing import List
from qdrant_client import QdrantClient
```

```python
documents:List[str]=[
    "What is retrieval augmented generation?",
    "Explain vector databases in simple terms.",
    "How does an embedding model work?",
    "What is the difference between LLMs and SLMs?",
    "How does vLLM improve inference performance?",
    "Explain cosine similarity.",
    "What is Qdrant used for?",
    "How do rerankers improve search quality?",
    "What is the role of KV cache in transformers?",
    "Explain agentic AI in one paragraph."
 ]
```

```python
inp_collection_name='just_collection' # table name in vector database 
```

```python
# DB creation
```

```python
client=QdrantClient(":memory:")
# client.upsert()
client.add(collection_name=inp_collection_name, documents=documents)
```

```python
client.get_collection(collection_name=inp_collection_name)
```
